#1. 오토글루온

오토글루온(AutoGluon)은 아마존에서 개발한 오픈소스 AutoML(자동 머신러닝) 프레임워크입니다.
오토글루온은 테이블, 이미지, 텍스트, 멀티모달 데이터에 대한 오토ML을 지원하며 본 코드에서는 테이블 데이터에 대한 실습을 진행해보도록 하겠습니다.

## 1. 기본 과정

```python
#오토글루온 설치
!pip install autogluon.tabular
```

```python
import pandas as pd
!wget https://raw.githubusercontent.com/shryu8902/KIRD_AUTOML/main/Titanic-Dataset.csv
titanic_data = pd.read_csv('./Titanic-Dataset.csv')
```

```python
from autogluon.tabular import TabularDataset, TabularPredictor

# TabularDataset으로 변환
train_data = TabularDataset(titanic_data)
# 예측하려는 컬럼 설정 (Survived)
label = 'Survived'
# 모델 학습
predictor = TabularPredictor(label=label).fit(train_data)
# 예측
predictions = predictor.predict(train_data)
# 성능 평가
performance = predictor.evaluate_predictions(y_true=train_data[label], y_pred=predictions)
print(performance)
```

## 2. 세부 내용 탐색

```python
# 예시를 위한 서브데이터셋 추출
train_data = TabularDataset(titanic_data)
subsample_size = 500
sampled_train_data = train_data.sample(n=subsample_size, random_state=0)
```

```python
print(train_data.shape)
print(sampled_train_data.shape)
print(train_data.head(3))
print(sampled_train_data.head(3))
```

```python
#predictor 생성 및 학습
predictor = TabularPredictor(label="Survived").fit(train_data)
```

```python
# 테스트용 데이터셋은 별개로 확보 필요 (pycaret과 차이)
test_data = train_data.drop(sampled_train_data.index)
```

```python
# fit 제외한 테스트 데이터
test_data.index
```

```python
# 학습된 데이터로 예측 수행
y_pred = predictor.predict(test_data)
y_pred.head(3)
```

```python
# 확률 파악시
y_pred_proba = predictor.predict_proba(test_data)
y_pred_proba.head()  # Prediction Probabilities
```

```python
# 성능평가 메트릭 계산
predictor.evaluate(test_data)
```

```python
# 모델리스트 확인하여 특정 모델로 선정 가능
predictor.model_names()
```

## 4. 프리셋

오토글루온에서는 다양한 수준의 프리셋을 제공함. 프리셋 설정을 통해서 모델 탐색 과정에서의 성능 수준을 조절 할 수 있음.
- best_quality : State-of-the-art (SOTA), much better than high_quality
- high_quality : Better than good_quality
- good_quality : Stronger than any other AutoML Framework
- medium_quality : Competitive with other top AutoML Frameworks


```python
time_limit = 30 # 초단위, 빠른 실습을 위한 설정
predictor_lv1 = TabularPredictor(label).fit(train_data, time_limit=time_limit, presets='best_quality')
predictor_lv2 = TabularPredictor(label).fit(train_data, time_limit=time_limit, presets='high_quality')
predictor_lv3 = TabularPredictor(label).fit(train_data, time_limit=time_limit, presets='good_quality')
predictor_lv4 = TabularPredictor(label).fit(train_data, time_limit=time_limit, presets='medium_quality')
```
