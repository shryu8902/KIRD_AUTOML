```python
import pandas as pd
!wget https://raw.githubusercontent.com/shryu8902/KIRD_AUTOML/main/Titanic-Dataset.csv
titanic_data = pd.read_csv('./Titanic-Dataset.csv')
```

```python
# 기본 동작 파이프라인
from pycaret.classification import ClassificationExperiment
# 분류모델 탐색 - 기본 셋팅
s = ClassificationExperiment()
s.setup(titanic_data, target = 'Survived') #실험 셋팅, data, target 인자는 필수
# compare_models 메서드를 통해 자동으로 실험을 수행하고 모델을 비교한다.
best = s.compare_models() #compare model로 모델 비교 수행
```

```python
#모델 평가 및 시각화
s.evaluate_model(best)
```

## 모델 학습 및 튜닝

### 다수의 모델을 동시에

```python
from pycaret.classification import ClassificationExperiment

s = ClassificationExperiment()
s.setup(titanic_data, target = 'Survived',
        ignore_features = ['PassengerId','Name','Ticket','Cabin'],
        fold=3, # cross validation
        use_gpu = True,
        session_id = 0) #random_state
```

```python
best = s.compare_models()
```

```python
best
```

```python
# hold-out testset 확인
s.test_transformed
```

```python
# hold-out test set에 대한 결과 예측 결과
s.predict_model(best)
```

```python
# hold-out test set에 대한 모델 평가
s.evaluate_model(best)
```

```python
# best model 예측 결과에 대한 accuracy
from sklearn.metrics import accuracy_score
accuracy_score(s.y_test_transformed, best.predict(s.X_test_transformed))
```

```python
best.score(s.X_test_transformed, s.y_test_transformed)
```

## 단일 모델 학습
개별적으로 모델을 설정하고 단일 모델에 대한 학습 및 최적화 수행 가능

```python
dt = s.create_model('dt') #결정트리 모델 생성
```

```python
#하이퍼 파라미터 최적화
tuned_dt = s.tune_model(dt, fold=3, n_iter = 20) # 결정트리 모델에 대한 파라미터 최적화
```

```python
# tuning 후 성능 증가 확인
print(dt.score(s.X_test_transformed, s.y_test_transformed))
print(tuned_dt.score(s.X_test_transformed, s.y_test_transformed))
```

## 앙상블 모델

```python
# 다수의 dt모델을 bagging 방법으로 앙상블 = 랜덤포레스트
bagging_dt = s.ensemble_model(dt, method = 'Bagging', n_estimators = 20)
```

```python
# 앙상블 모델과 단일 모델의 성능 비교
print(dt.score(s.X_test_transformed, s.y_test_transformed))
print(bagging_dt.score(s.X_test_transformed, s.y_test_transformed))
```

```python
# 단일 모델에 대한 앙상블이 아닌,
# 서로다른 모델에 대한 앙상블 모델
lr = s.create_model('lr')
knn = s.create_model('knn')
blended = s.blend_models([lr,knn,dt])
```

```python
# 개별 모델 및 앙상블 모델 성능 비교
print(dt.score(s.X_test_transformed, s.y_test_transformed))
print(lr.score(s.X_test_transformed, s.y_test_transformed))
print(knn.score(s.X_test_transformed, s.y_test_transformed))
print(blended.score(s.X_test_transformed, s.y_test_transformed))
```

```python
# compare model을 통한 상위 5개 모델 앙상블
blended_2 = s.blend_models(s.compare_models(n_select = 5))
#default : n_select = 1
```

```python
print(blended_2.score(s.X_test_transformed, s.y_test_transformed))
```

```python
s.predict_model(blended_2)
```
